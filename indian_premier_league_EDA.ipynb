### Most Successful IPL Teams 

```python
plt.figure(figsize=(50,60))
sns.factorplot(y='winner',kind='count',data=data1)
plt.show()
```

### Which season had most number of Matches

```python
sns.countplot(x='season',data=data1)
plt.show()
```

### Winner according to Season 

```python
teams_per_season=data1.groupby('season')['winner'].value_counts()
teams_per_season
```

### Winners of Team 1

```python
team1=data1.groupby('team1')['winner'].value_counts()
team1
```

### Winners of Team 2

```python
team2=data1.groupby('team2')['winner'].value_counts()
team2
```

### Win by Runs and Wickets Team 1

```python
data1.groupby('team1')[['win_by_runs','win_by_wickets']].sum().plot.bar(color=['red','blue'],figsize=(8,4))
plt.xlabel('Team1',fontsize=15)
plt.ylabel('Winner',fontsize=15)
plt.show()
```

### Win by Runs and Wickets Team 2

```python
data1.groupby('team2')[['win_by_runs','win_by_wickets']].sum().plot.bar(color=['red','blue'],figsize=(8,4))
plt.xlabel('Team2',fontsize=15)
plt.ylabel('Winner',fontsize=15)
plt.show()
```

### Player of Match

```python
playerofmatch=data1.groupby('player_of_match')['winner'].value_counts()
playerofmatch
```
